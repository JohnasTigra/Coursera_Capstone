````
# Scraping the Wikipedia Webpage:
web_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
web_request = requests.get(web_url)

soup = BeautifulSoup(web_request.content, 'html5lib')
neighbourhoods=[]
````

<h2>Fetching the Components</h2>
<p> In the cell below, I fetch the necessary neighbourhood data from the HTML code and print the list containning all the neighbourhood dictionaries. In the printout I see many unassigned values, they need to be dropped.</p>

````
table = soup.find('table', attrs = {'class': 'wikitable sortable'})
soup.prettify()
for row in table.findAll('tr')[1:]: 
  col = row.findAll('td')
  neighbour={}
  neighbour['Neighbourhood'] = str(col[2].get_text())
  neighbour['Borough'] = str(col[1].get_text())
  neighbour['Pincode'] = str(col[0].get_text())
  neighbourhoods.append(neighbour)
  print(neighbourhoods)    
````

<h2> Dropping the unassigned values. </h2>
<p> Here I drop the unassigned values from the dataset and create a clean dataset. But here I see an encoding error. In the <b>Neighbourhood</b> column there are extra <b> \n </b> added, which needs to be eliminated. </p>

````
neighbourhoods = pd.DataFrame(neighbourhoods)
neighbourhoods.index.values
# list1 = neighbourhoods.iloc[0]
# list1['Neighbourhood'] = ['go']
# str(list1['Borough']+'\n')
````

<h2> Eliminating the encoding error and assigning the Unassigned Neighbourhood.</h2>
<p> Here I use a python for iteration to replace the certain charcters to obtain the clean dataset. Next, I again an iteration technique to assign the borough name to the unassigned neighbouhood, whereas applicable.</p>

````
b = '\n'
values = []
for value in neighbourhoods['Neighbourhood'].values:
  for f in b:
    value = value.replace(f, "")
    values.append(value)
print(values)
neighbourhoods['Neighbourhood'] = values

neighbourhoods = pd.DataFrame(neighbourhoods)

index_values = neighbourhoods.index.values
for i in index_values:
  select_row = neighbourhoods.iloc[i]
  if select_row['Borough'] != 'Not assigned':
    if select_row['Neighbourhood'] == 'Not assigned':
      neighbourhoods.loc[i, 'Neighbourhood'] = select_row['Borough']
neighbourhoods   
````

<h2> Cleaning the Dataset</h2>
<p> Here, I assign the 'Not Assigned' values to NaN and drop the rows with complete NaN values to obtain the cleanest form of the data. </p>

````
neighbourhoods = pd.DataFrame(neighbourhoods)
neighbourhoods.replace('Not assigned', np.nan, inplace = True)
neighbourhoods.dropna(axis = 0, inplace = True)
neighbourhoods
````

<h2> Observing Uniqueness in the datset </h2>
<p> Here let's take a look at the unique values in the dataset in accordance with the borough value. </p>

````
final_df = neighbourhoods[['Pincode', 'Borough', 'Neighbourhood']]
print('Observation 1: There are {} unique Boroughs'.format(len(final_df['Borough'].unique())))
````

<h2> Obtainning the Final Dataframe. </h2>
<p> Here I use the groupby method to sort the table with Pincodes and join the Neighbourhoods falling into the same pincode. </p>

````
fd_grouped = final_df.groupby('Pincode')['Neighbourhood', 'Borough'].agg(lambda x: ', '.join(set(x))).reset_index()
fd_grouped
````

<h2> Obtainning the Shape of the Dataframe</h2>

````
the_final_dataframe = pd.DataFrame(fd_grouped)
print("The Shape of the Final Dataframe is : ", the_final_dataframe.shape)
````
